### This file deal with all the basic structure fix of dataset
### Like droping unnecessary column, adding extra column as needed and concating all the dataset

```python
import pandas as pd
import os
```

```python
def drop_last_six_columns(file_path):
    df = pd.read_csv(file_path)
    
    # drop unnecessary columns
    df = df.iloc[:, :-7]
    
    # write back to csv
    df.to_csv(file_path, index=False)
```

```python

def add_state_and_constituency(file_path):
    df = pd.read_csv(file_path)
    
    # get state and constituency name from the file name
    file_name = os.path.basename(file_path)
    state = file_name.split('_')[0]
    constituency_name = file_name.split('_')[1].split('.')[0]
    
    # Add the new columns to the DataFrame
    df['State'] = state
    df['Constituency Name'] = constituency_name
    
    # Save the modified DataFrame back to the CSV file without an extra index column
    df.to_csv(file_path, index=False)
```

```python

def process_files_in_directory(directory_path):
    # Iterate over all files in the directory and subdirectories
    for root, dirs, files in os.walk(directory_path):
        for file in files:
            if file.endswith('.csv'):
                file_path = os.path.join(root, file)
                drop_last_six_columns(file_path)
                add_state_and_constituency(file_path)
                print(f"Processed file: {file_path}")

# Specify the main directory containing the subfolders with CSV files
main_directory_path = 'data'

# Process all CSV files in the main directory and its subdirectories
process_files_in_directory(main_directory_path)
```
